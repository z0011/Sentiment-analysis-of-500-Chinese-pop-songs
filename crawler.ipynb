```python
import requests
import urllib.parse
import re
from bs4 import BeautifulSoup

url_rt = 'http://mojim.com/'

def get_lyric(url):
    url = urllib.parse.urljoin(url_rt, url)
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'lxml')
    lyric = soup.find('dl', 'fsZx1')
    a = re.compile('^\[\d+')
    
    lyric_list = list()
    for s in lyric.stripped_strings:
        if '：'in s:
            continue
        if '['in s:
            continue
        if s == '更多更詳盡歌詞 在' or s == '※ Mojim.com　魔鏡歌詞網':
            continue
        if a.match(s):
            break
        if s == '感謝':
            break
        s = re.sub(u'[\u3000,\xa0]',u'',s) 
        lyric_list.append(s)
    singer = lyric_list.pop(0)
    name = lyric_list.pop(0)
    L = get_ytlk(singer,name)
    song_detail = {
        'singer':singer,
        'name':name,
        'tylk':L,
        'lyric':lyric_list
    }
    return song_detail
def get_ytlk(singer,name):
    url = "https://www.youtube.com/results?search_query="+singer+name
    request = requests.get(url)
    content = request.content
    soup = BeautifulSoup(content, "html.parser")
    x=soup.select(".yt-lockup-video")[0]
    d=x.select("a[rel='spf-prefetch']")
    Title=(d[0].get("title"))
    Link=("https://www.youtube.com{}".format(d[0].get("href")))
    L=[Title,Link]
    return L
```

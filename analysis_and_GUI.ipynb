```python
#read CVAW - The Chinese Valence-Arousal Word
import csv
import numpy as np
L_arousal=[]
L_a=[]
f = open('cvaw3.csv', 'r',encoding = 'utf8')
for row in csv.reader(f):
    L_arousal.append(row)
f.close()

#build training data of arousal
for i in range(1,3553):
    L_a.extend([[L_arousal[i][1],eval(L_arousal[i][4])]])
m=[]
for i in range(3552):
    m.append(L_a[i][1])
meann=np.mean(m)
h_a=[]
l_a=[]
for i in range(3552):
    if L_a[i][1]>meann:
        h_a.append(L_a[i][0])
    else:
        l_a.append(L_a[i][0])
```

```python
#save training data of arousal
f_h_a = open('h_a.txt', 'w',encoding = 'utf8')
f_h_a.write('，'.join(h_a))
f_h_a.close()

f_l_a = open('l_a.txt', 'w',encoding = 'utf8')
f_l_a.write('，'.join(l_a))
f_l_a.close()
```

```python
#sentiment analysis of valence with buit-in data of SnowNlp 
import os
from snownlp import SnowNLP
os.listdir('lyric')
singer_list=os.listdir('lyric')
singer_list_detail=[0]*20
for i in range(20):
    f=open(singer_list[i], 'r',encoding='UTF-8')
    singer_list_detail[i]=eval(f.read())
# singer_list_detail[0~20][0~25] is a dict with 'singer', 'name', 'tylk'([url_title,url]),'lyric'

Lx=[[{'singer':0,'song_name':0,'url':0,'sen_pt':0} for i in range(25)] for j in range(20)]

for i in range(20):
    for j in range(25):
        s=','.join(singer_list_detail[i][j]['lyric'])
        sen = SnowNLP(s)
        sen_pt=sen.sentiments
        Lx[i][j]['singer']=singer_list_detail[i][j]['singer']
        Lx[i][j]['song_name']=singer_list_detail[i][j]['name']
        Lx[i][j]['url']=singer_list_detail[i][j]['tylk'][1]
        Lx[i][j]['sen_pt']=sen_pt
        
        
print(Lx)
```

```python
F_x = open('F_x_valence.txt', 'w',encoding='UTF-8')
F_x.write(str(Lx))
F_x.close()
```
